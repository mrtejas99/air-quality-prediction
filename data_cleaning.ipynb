### Panaji



```python
df_panaji = pd.read_csv("dataset\\panaji.csv", parse_dates=['SAMPLING DATE'])
df_panaji.head()
```

```python
df_panaji['SAMPLING DATE'] = pd.to_datetime(df_panaji['SAMPLING DATE'], errors='coerce',infer_datetime_format=True, dayfirst=True, format='%d%m%Y').dt.strftime('%d/%m/%Y')
df_panaji = df_panaji.astype({'SAMPLING DATE':'datetime64[ns]'})
```

```python
df_panaji.info()
```

```python
df_panaji.drop('AQI REMARKS', axis=1, inplace=True)
df_panaji.info()
```

```python
df_panaji.head()
```

```python
import numpy as np
df_panaji['SO2 µg/m3'] = df_panaji['SO2 µg/m3'].replace('BDL', np.nan)
df_panaji['NO2 µg/m3'] = df_panaji['NO2 µg/m3'].replace('BDL', np.nan)

df_panaji = df_panaji.replace({'PM2.5 µg/m3':{
      'BDL':np.nan,
      'FPS not working':np.nan,
      'FPS not operated due to rains':np.nan,
      'not operated due to rains':np.nan,
      '$':np.nan,
      '*':np.nan,
      '/':np.nan,
      'Monitoring stopped due to Rainy Season':np.nan
      }})
# df_panaji = df_panaji.astype({'SAMPLING DATE':'datetime64', 'SO2 µg/m3':'float','NO2 µg/m3':'float','PM10 µg/m3':'float','PM2.5 µg/m3':'float', 'AQI':'float'})
df_panaji = df_panaji.astype({'SO2 µg/m3':'float', 'NO2 µg/m3':'float', 'PM2.5 µg/m3':'float', 'AQI':'float'})
df_panaji = df_panaji.set_index('SAMPLING DATE')
df_panaji.info()
```

```python
import matplotlib.pyplot as plt
plt.style.use('default')
%matplotlib inline
```

```python
df_vasco['City'] = "Vasco"
df_margao['City'] = "Margao"
df_ponda['City'] = "Ponda"
df_mapusa['City'] = "Mapusa"
df_panaji['City'] = "Panaji"
```

```python
print(df_vasco.columns)
print(df_panaji.columns)
print(df_margao.columns)
print(df_ponda.columns)
print(df_mapusa.columns)
```

```python
df_vasco.rename(columns={'SO2 µgm3':'SO2 µg/m3', 'NO2 µgm3':'NO2 µg/m3', 'PM10 µgm3 ':'PM10 µg/m3', 'PM2.5 µgm3 ':'PM2.5 µg/m3', 'AQI':'AQI' }, inplace=True)
df_panaji.rename(columns={'SO2 µgm3':'SO2 µg/m3', 'NO2 µgm3':'NO2 µg/m3', 'PM10 µgm3 ':'PM10 µg/m3', 'PM2.5 µgm3 ':'PM2.5 µg/m3', 'AQI':'AQI' }, inplace=True)
df_margao.rename(columns={'SO2 µg/m3':'SO2 µg/m3', 'NO2 µg/m3':'NO2 µg/m3', 'PM10 µg/m3 ':'PM10 µg/m3', 'PM2.5 µg/m3 ':'PM2.5 µg/m3', 'AQI':'AQI' }, inplace=True)
df_ponda.rename(columns={'SO2 µg/m3':'SO2 µg/m3', 'NO2 µg/m3':'NO2 µg/m3', 'PM10 µg/m3 ':'PM10 µg/m3', 'PM2.5 µg/m3 ':'PM2.5 µg/m3', 'AQI':'AQI' }, inplace=True)
df_mapusa.rename(columns={'SO2 µg/m3':'SO2 µg/m3', 'NO2 µg/m3':'NO2 µg/m3', 'PM10 µg/m3':'PM10 µg/m3', 'PM2.5 µg/m3':'PM2.5 µg/m3', 'AQI':'AQI' }, inplace=True)
```

```python
df_total = pd.concat([df_vasco, df_panaji, df_margao, df_ponda, df_mapusa], axis=0)
#df_total = pd.concat([df_vasco, df_panaji, df_margao,  df_mapusa], axis=0)


#fix for extra pm10 column-confirm
from IPython.display import display, HTML
display(HTML(df_total.to_html()))
```

```python
#df_total['PM10 µg/m3'].unique()
```

```python
pollutants = ['PM2.5 µg/m3', 'PM10 µg/m3', 'NO2 µg/m3', 'SO2 µg/m3']

temp=df_total.groupby('SAMPLING DATE')['PM2.5 µg/m3', 'PM10 µg/m3', 'NO2 µg/m3', 'SO2 µg/m3'].sum().reset_index()
temp=temp.melt(id_vars="SAMPLING DATE",value_vars=pollutants,var_name='Pollutants',value_name='Count')
temp.head()

import plotly.express as px
fig=px.area(temp,x='SAMPLING DATE',y='Count',color='Pollutants',height=600,title='Pollutants over time',color_discrete_sequence=['#393e46', '#ff2e63', '#456fe3','#78ffee'])
fig.update_layout(xaxis_rangeslider_visible=True)
fig.show()
```
